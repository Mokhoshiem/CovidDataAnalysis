# Q2:
## Countries with at least 8 hospital beds per thousand are aged?

```python
#  Droping nan values
df.dropna(inplace=True)
```

```python
df.isna()
```

```python
df[non_null_values]
```

```python
#  Countries with 8 beds at least
df_eight_beds = df.query('hospital_beds_per_thousand >= 8')
df_eight_beds.head(2)
```

```python
df_less_than_eight_beds = df.query('hospital_beds_per_thousand < 8')
df_less_than_eight_beds.head(2)
```

```python
mean_aged_65_per_eight_beds = df_eight_beds['aged_65_older_percent'].mean()
mean_aged_65_per_eight_beds
```

```python
mean_aged_65_per_less_than_eight_beds = df_less_than_eight_beds['aged_65_older_percent'].mean()
mean_aged_65_per_less_than_eight_beds
```

## It's noticed that countries with more aged percentage have more beds
#### Countries with 8 or more hospital beds have about 18% of the population aged more than 65 years old, while countries with 8 or less hospital beds have about 10% of thier population aged more than 65 years old.
# There for we might say the lack of hospital services may lead to shorter lifetime.

```python
# importing sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
plt.scatter(df['hospital_beds_per_thousand'],df['aged_65_older_percent'])
plt.ylabel('Aged 65 of population %')
plt.xlabel('hospital beds per thousands')
plt.title('relation between health care and lifetime')
plt.show()
```

```python
# training ou model
x_train, x_test, y_train, y_test = train_test_split(df['hospital_beds_per_thousand'], df['aged_65_older_percent'])
```

```python
lr = LinearRegression()
lr.fit(x_train.values.reshape(-1,1), y_train.values)
```

```python
prediction = lr.predict(x_test.values.reshape(-1,1))
```

```python
plt.plot(x_test, prediction, label='Linear Regression', color='b')
plt.scatter(x_test, y_test, label='Actual Data', color='g')
plt.legend()
plt.show()
```

```python
# Score the test
lr.score(x_test.values.reshape(-1,1), y_test.values)
```

# Q5:
## How likely an individual in Egypt can survive the infection?

```python
# survival rate = 1 - (total deaths / total infections)
total_deaths_EG = country_groups['new_deaths']['Egypt']
total_infections_EG = country_groups['new_cases']['Egypt']
death_rate_EG = total_deaths_EG/total_infections_EG
```

```python
death_rate_EG
```

```python
# Surviving rate = 1 - death rate
surviving_rate_EG = 1 - death_rate_EG
surviving_rate_EG
```

## An individual in Egypt is likely to survive the infection with a 69% survival rate
